### Blocos identados e aninhamento

Cada bloco de um comando IF pode conter um novo conjunto de instruções onde novos blocos e identações podem ser necessários. No exemplo abaixo mostraremos como um IF pode ser necessário dentro de outro IF.

```python
numero = input("Digite um numero:")
if int(numero)%2==1:
  print('{} é ímpar! Gostaria de alterar para o próximo número par?'.format(numero))
  resp = input('Responda S para SIM ou N para Não')
  if resp == 'S':
    numero = str(int(numero)+1)
else:
  print('{} é par!'.format(numero))
print("Resultado final: {}".format(numero))
```

#### Considerações importantes:

Note na célula acima os seguintes detalhes:

- o comando IF interno ao bloco verdadeiro do primeiro IF requer um novo bloco verdadeiro, num segundo nível de identação à direita, onde a variável *numero* será incrementada para se tornar par

- esse segundo IF não tem clausula ELSE

- o incremento de *numero* requereu duas operações de mudança de tipo. A primeira transforma a variável de string para inteiro, para permitir a operação algébrica, e a segunda retorna a variável *numero* para seu tipo original string. Essa estratégia de uso de uma função dentro da outra é chamada de **aninhamento**. A função dentro do parênteses é executada primeiro e seu resultado é usado como entrada para a função externa.

## Loops: o comando WHILE

Note que no exercício 5, mesmo com um teste para saber se o usuário digitou o mesmo número duas vezes e pedir que ele mude o segundo número informado, ele ainda pode burlar a regra e redigitar o mesmo número um terceira vez. Como evitar que isso aconteça?

Uma solução para isso seria ter uma sequência de operações mais ou menos assim:

- leia o primeiro número
- leia o segundo número
- enquanto o primeiro número for igual ao segundo faça:
   - avise o usuário que deve mudar o segundo número
   - leia o segundo número

Essa parte do algoritmo que se repete enquanto a condição de parada não é satisfeita é chamada de LOOP.

O algoritmo acima em Python seria:

```python
numero1 = input("Digite um número:")
numero2 = input("Digite um segundo número:")
while numero1 == numero2:
  print('Não repita o mesmo número!')
  numero2 = input("Digite um segundo número diferente do primeiro:")
print('{} é diferente de {}'.format(numero1,numero2))
```

## Loops: o comando FOR

Em algumas situações deseja-se que um bloco de código seja executado um número específico de vezes, e não até que uma condição de parada seja satisfeita.

Nesse caso deve-se usar o comando FOR, que estabelece o número de repetições.

```python
numero = input("Digite um número:")
for i in range(0,int(numero)):
  print(i,int(numero)-i)
```

#### Considerações importantes:

Note na célula acima os seguintes detalhes:

- a variável i (qualquer nome pode ser usado) controla o loop. Seu valor vai variar segundo definido na função *range*, que define o início e o final de um intervalo, sendo que i vai começar valendo o valor indicado no primeiro parâmetro do range, vai ser incrementado de 1 a cada execução do loop, e o loop vai parar quando i atingir o segundo valor indicado no range. Note que quando i atinge esse limite o bloco não é executado. No caso do exemplo quando i chegou a 10 o loop parou.

- a função range precisa receber dois números inteiros como parâmetros

- o bloco de comandos do loop precisa estar deslocado à direita com relação ao FOR

#### Interrompendo um FOR

O comando FOR pode ser interrompido antes do número previsto de loops usando a função BREAK. Dessa forma o range define o intervalo máximo de execuções mas o break permite parar antes caso uma condição específica tenha sido atingida. Exemplo:

```python
acumulador = 0
numero = input("Digite um número:")
for i in range(0,int(numero)):
  acumulador += i
  print(i,int(numero)-i,acumulador)
  if acumulador > int(numero):
    break
```

#### Considerações importantes:

Note a operação algébrica dentro do loop: acumulador += i. Ela é equivalente a acumulador = acumulador + i

A função range aceita outras opções de entrada:

range(fim)   
range(inicio,fim,passo)

```python
for i in range(10):
  print(i)
```

```python
for i in range(10,20,2):
  print(i)
```

## Exercício 7:

Adapte seu algoritmo do jogo de adivinhação de forma a escolher o número desejado usando a função randrange e limitando o número de tentativas a um valor pré-definido no programa. A cada chute do usuário avise quantos chutes ele ainda tem.

Envie sua solução no AVA na tarefa designada.Envie somente as linhas de código da sua solução.

```python
import random

PRE_FIXED = random.randint(0, 10)

attempts = 3

while attempts > 0:
    bet = int(input("Please, enter a number: "))  
    if bet == PRE_FIXED:
        print("The secret number is: {}".format(PRE_FIXED))
        print("Congratulations, you got the number right!")
        break
    elif bet > PRE_FIXED:
        attempts -= 1
        print("You have {} attempts left".format(attempts))
        print("Please enter a smaller number.")
    elif bet < PRE_FIXED:
        attempts -= 1
        print("You have {} attempts left".format(attempts))
        print("Please enter a bigger number.")
    
if attempts == 0:
    print("Sorry, you've used all your attempts. The secret number was: {}".format(PRE_FIXED))
```
